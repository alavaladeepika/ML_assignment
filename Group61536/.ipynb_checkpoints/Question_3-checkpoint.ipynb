```python
%matplotlib inline      
import matplotlib.pyplot as plt
data.hist(bins=50, figsize=(20,15))
plt.show()
#Observations:
#F10, F11, F6, F7, F8 seem to be tail heavy since they extend much farther to the right of the median than to the left.
#This makes detection of patterns difficult. Desired shape : bell-shaped distribution
```

```python
corr_matrix = data.corr()
corr_matrix
#The correlation matrix gives the relation between each of the attributes. 
#The correlation coefficient ranges from 1 to 1. When it is close to 1, it means that there is a strong positive correlation
#When the coefficient is close to 1, it means that there is a strong negative correlation
#Coefficients close to zero mean that there is no linear correlation
#For example, the correlation coefficient(F0,F2) is close to 1 which implies F0 increases linearly with an increase in F2 and vice-versa'''
```

```python
from pandas.plotting import scatter_matrix
attributes = ["F0","F1","F2","F3","F4","F5","F6","F7","F8","F9","F10","F11"]
scatter_matrix(data[attributes], figsize=(12, 8))
#Scroll down to view the plots.
#There seems to be a linear correlation between some of the attributes. Other correlations seem to be non linear and no correlation at all
```
